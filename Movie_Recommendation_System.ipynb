```python
a=dataset
a=a.groupby('title')['rating'].mean()
a
```

```python
sorted_rating_wise_movie=a.sort_values(ascending=False)
sorted_rating_wise_movie
```

```python
b=dataset.groupby('title')['rating'].count()
b
```

```python
new_record=pd.DataFrame()
new_record['Average_ratings']=a
new_record['Count of total ratings']=b
new_record
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_style('white')
%matplotlib inline
```

```python
new_record.columns
```

```python
plt.figure(figsize=(10,9))
new_record['Count of total ratings'].hist(bins=100)
```

```python
plt.figure(figsize=(10,4))
new_record['Average_ratings'].hist(bins=70)
```

```python
plt.figure(figsize=(8,6))
sns.jointplot(x='Average_ratings',y='Count of total ratings',data=new_record,alpha=0.4)
```

```python
moviematrix=dataset.pivot_table(index='userId',columns='title',values='rating')
```

```python
moviematrix.head()
```

```python
moviematrix.tail()
```

```python
moviematrix.columns
```

```python
moviematrix['xXx (2002)']
```

```python
xXx_2002_ratings=moviematrix['xXx (2002)']
```

```python
xXx_2002_ratings.head()
```

```python
xXx_2002_ratings.tail()
```

```python
movies_similar_to_xxx_2002=moviematrix.corrwith(xXx_2002_ratings)
movies_similar_to_xxx_2002
```

```python
import matplotlib.pyplot as plt
correlation_metrics=dataset.corr()
fig=plt.figure(figsize=(14,9))
sns.heatmap(correlation_metrics,annot=True,vmax=0.9,square=True)
plt.show()
```

```python
xXx_2002_corr=pd.DataFrame(movies_similar_to_xxx_2002,columns=['Correlation'])
xXx_2002_corr.dropna(inplace=True)
xXx_2002_corr
```
